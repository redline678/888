## Pandas - фильтрация данных

Считать данные по  [адресу](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 
Присвоить переменной chipo

```python
chipo = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
chipo
```

Как много продуктов стоят более чем $10.00?

```python
chipo['price_modified'] = chipo['item_price'].apply(lambda x: x.replace('$', ''))
chipo['price_modified'] = chipo['price_modified'].apply(lambda x: x.replace(' ', ''))
chipo['price_modified'] = pd.to_numeric(chipo['price_modified'])
chipo
```

```python
chipo['choice_description'] = chipo['choice_description'].fillna('Empty')
chipo['item_item'] = chipo['item_name'].map(str) +' with ' + chipo['choice_description']
chipo
```

```python
chipo['corrected_price'] = chipo['price_modified'] / chipo['quantity']
chipo
```

```python
corrected_price = chipo.loc[chipo['corrected_price'] > 10]
corrected_price['item_item'].count()
```

# Как много продуктов стоят более чем $10.00? - ОТВЕТ: 1034

Какова цена каждого предмета? Напечатать столбцы item_name и item_price

```python
b = chipo[['item_item','corrected_price']]
b.drop_duplicates(inplace = True)
b
```

Отсортировать по имени предмета

```python
b.sort_values('item_item')
```

Сколько было заказано самого дорогого предмета?

```python
max_price = max(b['corrected_price'])
```

```python
b.loc[b['corrected_price'] == max_price].count()
```

Сколько раз был заказан предмет "Veggie Salad Bowl"?

```python
vsb = chipo.loc[chipo['item_name'] == "Veggie Salad Bowl"]
vsb.count()
```
