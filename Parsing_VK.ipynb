```python
import requests
import pandas as pd
```

```python
owner_id = -41653980
offset = 0
all_posts=[]
while True:
    r = requests.get('https://api.vk.com/method/wall.get', params = {'owner_id': owner_id,
                                                                 'access_token': '',
                                                                 'v': 5.52,
                                                                 'count':1000,
                                                                 'offset': offset})
    post = r.json()#['response']
    all_posts.append(post)
    offset += 100
    if offset == 1000:
        break

    
    #r.json()
```

```python
df1 = pd.io.json.json_normalize(all_posts)
df1
```

```python
#df1 = pd.io.json.json_normalize(r.json())
cz = df1['response.items']
cz
```

```python
data = pd.DataFrame()
```

```python
for index in range(0,len(cz)):
    a = cz[index]
    l = len(a) # всегда 100
    dates = []
    
    for i in range(0,l):
        b = a[i]
        dates.append(b['date']) #вытаскиваем даты постов
    #
    likes = []
    for i in range(0,l):
        b = a[i]
        likes.append(b['likes'])
    likes_count = []  
    for i in range(0,l):
        c = likes[i]
        likes_count.append(c['count']) #Вытаскиваем лайки
    #
    com = []
    for i in range(0,l):
        b = a[i]
        com.append(b['comments'])
    com_count = []  
    for i in range(0,l):
        c = com[i]
        com_count.append(c['count'])#вытаскиваем комментарии
    #
    rep = []
    for i in range(0,l):
        b = a[i]
        rep.append(b['reposts'])
    rep_count = []  
    for i in range(0,l):
        c = rep[i]
        rep_count.append(c['count'])#вытаскиваем репосты
            #вытаскиваем 'marked_as_ads'
    ad = []
    for i in range(0,l):
        b = a[i]
        ad.append(b['marked_as_ads'])
    df = pd.DataFrame([dates,likes_count,com_count,rep_count,ad])
    df = df.transpose()
    data = pd.concat([data, df])
    
data
```

```python
data.rename(columns={0: 'Date', 1: 'Likes', 2: 'Comments', 3: 'Reposts', 4: 'Ad'}, inplace=True)
data
```
